# Weibull Life Data Analysis

```python
import os

%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
from lifelines import WeibullFitter
from lifelines.plotting import qq_plot
```

```python
df = pd.read_csv(os.path.join('source', 'data', 'data4.csv'))
```

```python
df.head()
```

```python
df.info()
```

```python
wbf = WeibullFitter()
wbf.fit(durations=df['Time'], event_observed=df['Event'], timeline=np.arange(0, 10000, 500), ci_labels=['lower .95', 'upper .95'])
```

```python
ax0 = wbf.plot()
ax0.legend(['cumulative hazard'], loc='upper left')
```

```python
wbf.confidence_interval_.plot()
# ax1 = wbf.confidence_interval_.plot()
# ax1.set_ylim(-.05, .75)
# ax1.legend(['lower .95', 'upper .95'], loc='upper left')
```

```python
ax2 = wbf.cumulative_density_.plot()
#ax2.set_ylim(-.05, .75)
ax2.legend(['cumulative hazard'], loc='upper left')
```

```python
wbf.hazard_.plot()
```

```python
wbf.survival_function_.plot()
```

```python
wbf.survival_function_
```

```python
wbf.confidence_interval_survival_function_
```

```python
surv_df = wbf.confidence_interval_survival_function_.join(wbf.survival_function_.rename(columns = {'Weibull_estimate': 'reliability'}))
ax3 = surv_df['reliability'].plot(figsize=(10, 7), title='reliability over time', legend=True)
ax3.set_xlabel('time')
ax3.set_ylabel('reliability')
ax3.fill_between(surv_df.index, surv_df['lower .95'], surv_df['upper .95'], facecolor='grey', alpha=.5)
```
